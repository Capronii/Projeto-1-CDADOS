### Montando um Classificador Naive-Bayes

```python
train['Classifica√ß√£o'] = train['Classifica√ß√£o'].astype('category')
train['Classifica√ß√£o'].cat.categories = ['Positivo','Negativo','Coment√°rio fora de contexto'] 
test['Classifica√ß√£o']=test['Classifica√ß√£o'].astype('category')
test['Classifica√ß√£o'].cat.categories= ['Positivo','Negativo','Coment√°rio fora de contexto'] 
train.head(5)
```

```python
import re

pos=train.loc[train['Classifica√ß√£o']=='Positivo']
pos1 = ' '.join(pos.Treinamento)

def cleanup(text):
    punctuation = '[!-.:?;,@]'
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    text_subbed=text_subbed.replace('√°','a')
    text_subbed=text_subbed.replace('√£','a')
    text_subbed=text_subbed.replace('√¢','a')
    text_subbed=text_subbed.replace('√©','e')
    text_subbed=text_subbed.replace('√™','e')
    text_subbed=text_subbed.replace('√≠','i')
    text_subbed=text_subbed.replace('√¥','o')
    text_subbed=text_subbed.replace('√≥','o')
    text_subbed=text_subbed.replace('√∫','u')
    text_subbed=text_subbed.replace('√ß','c')
    return text_subbed

cleanup(pos1)
lista_p= pos1.split(' ')

df_positivo=pd.Series(lista_p)

abs_positivo=df_positivo.value_counts()
rel_positivo=df_positivo.value_counts(True)

print(abs_positivo)
print(rel_positivo)
```

### Verificando a performance do Classificador

Agora voc√™ deve testar o seu classificador com a base de Testes.

```python
import re

tudo= ' '.join(train.Treinamento)

def cleanup(text):
    punctuation = '[!-.:?;]'
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    text_subbed=text_subbed.replace('√°','a')
    text_subbed=text_subbed.replace('√£','a')
    text_subbed=text_subbed.replace('√¢','a')
    text_subbed=text_subbed.replace('√©','e')
    text_subbed=text_subbed.replace('√™','e')
    text_subbed=text_subbed.replace('√≠','i')
    text_subbed=text_subbed.replace('√¥','o')
    text_subbed=text_subbed.replace('√≥','o')
    text_subbed=text_subbed.replace('√∫','u')
    text_subbed=text_subbed.replace('√ß','c')
    return text_subbed

cleanup(tudo)
lista_t=tudo.split(' ')

df_t=pd.Series(lista_t)

abs_t=df_t.value_counts()
rel_t=df_t.value_counts(True)

print(abs_t)
print(rel_t)
```

```python
total_t=len(abs_t)
total_p=len(abs_positivo)
total_n=len(abs_negativo)
total_c=len(abs_comentario)

```

```python
prob_p=[]
prob_n=[]
prob_c=[]

classify=[]
proba_p=1
proba_n=1
proba_c=1
for j in test['Teste']:
    limpando=cleanup(j)
    lista_t=limpando.split()
    proba_p=1
    proba_n=1
    proba_c=1
    for i in lista_t:
        if i in abs_positivo:
            proba_p *= ((abs_positivo[i]+1)/(total_t+total_p))
        if i not in abs_positivo:
            proba_p *= ((1)/(total_t+total_p))
        if i in abs_negativo:
            proba_n *= ((abs_negativo[i]+1)/(total_t+total_n))
        if i not in abs_negativo:
            proba_n *= ((1)/(total_t+total_n))
        if i in abs_comentario:
            proba_c *= ((abs_comentario[i]+1)/(total_t+total_c))
        if i not in abs_comentario:
            proba_c *= ((1)/(total_t+total_c))

    prob_p.append(proba_p)
    prob_n.append(proba_n)
    prob_c.append(proba_c)


for i in range(0,len(prob_p)):
    if (prob_p[i]>prob_n[i]) and (prob_p[i]>prob_c[i]):
        classify.append("Positivo")
    elif (prob_n[i]>prob_p[i]) and (prob_n[i]>prob_c[i]) :
        classify.append("Negativo")
    else:
        classify.append("Comentario fora do contexto")


test['Classifica√ß√£o final']=classify
test
```

```python
pd.crosstab(test['Classifica√ß√£o'],test['Classifica√ß√£o final'],normalize='columns').round(4)*100
```
